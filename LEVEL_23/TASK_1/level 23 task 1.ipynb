```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'life.csv'
df = pd.read_csv(file_path)
```

```python
avg_life_expectancy = df.drop(columns='country').mean()
```

```python
years = avg_life_expectancy.index.astype(int)
values = avg_life_expectancy.values
```

```python
# 4. Plot
plt.figure(figsize=(12, 6))
plt.plot(
    years,
    values,
    marker='o',        
    linestyle='-',     
    linewidth=2,
    markersize=6
)
plt.title('Average Life Expectancy Over Time')
plt.xlabel('Year')
plt.ylabel('Average Life Expectancy (Years)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
