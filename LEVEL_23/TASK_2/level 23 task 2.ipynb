```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('incomes.csv')
```

```python
stats = df.drop(columns='country') \
          .agg(['mean', 'min', 'max'], axis=0) \
          .T 
```

```python
stats.columns = ['Average', 'Min', 'Max']
```

```python
years = stats.index.astype(int)
```

```python
plt.figure(figsize=(12,6))

plt.plot(years, stats['Average'], label='Average Income', linewidth=2, marker='o')
plt.plot(years, stats['Min'],     label='Min Income',     linewidth=2, linestyle='--', marker='v')
plt.plot(years, stats['Max'],     label='Max Income',     linewidth=2, linestyle='-.', marker='^')
tick_locs = years[::10]
plt.xticks(tick_locs, rotation=45)

plt.title('Income Statistics by Year')
plt.xlabel('Year')
plt.ylabel('Income')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

```
