```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cell_phones.csv')
```

```python
print(df.head())
print(df.dtypes)
```

```python

year_cols = [c for c in df.columns if c != 'country']
```

```python

df[year_cols] = df[year_cols].apply(pd.to_numeric, errors='coerce')
```

```python

avg_per_100 = df[year_cols].mean(axis=0, numeric_only=True)
```

```python
avg_per_person = avg_per_100 / 100.0
years = avg_per_person.index.astype(int)
phones_per_person = avg_per_person.values
```

```python
plt.figure(figsize=(12,6))
plt.plot(
    years,
    phones_per_person,
    marker='o',
    linestyle='-',
    linewidth=2,
    markersize=6,
    color='tab:green'
)
plt.title('Average Number of Cell Phones per Person Over Time')
plt.xlabel('Year')
plt.ylabel('Cell Phones per Person')
plt.grid(alpha=0.3)
plt.xticks(years[::10], rotation=45)
plt.tight_layout()
plt.show()
```
